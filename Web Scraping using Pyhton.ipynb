#### WEB SCRAPING

<b>Definition</b><br>
<ul>
    <li>The Process of extracting html data from webpage / website.</li>
    <li>Transforming HTML unstructured data to structure data into excel or dataset.</li>
    <li>Let's us study this concept with an example of extracting name of the all the services offered on <a href="www.eshikshak.co.in">http://www.ehsikshak.co.in</a> website.</li>

#### Step 1

To start with web scraping, we need two libraries: <code>BeautifulSoup</code> in <code>bs4</code> and request in <code>urllib</code>. Import both of these python packages.

```python
#import packages(libraries)
from bs4 import BeautifulSoup
import urllib.request
```

#### Step 2

Select the url, for extracting its HTML elements.

```python
#target URL 
url = "http://www.eshikshak.co.in"
```

#### Step 3

We could access the content on this webpage and save the HTML in “myUrl” by using <code>urlopen()</code> function in request.

```python
#use request to open URL
myUrl = urllib.request.urlopen(url)
```

#### Step 4

Create an <b>object</b> of <code>BeautifulSoup</code> to further extract the webpage element data, using its various inbuilt functions.

```python
# soup is an object of BeautifulSoup that will allow us to used all its inbuild functions to extract webpage element data
soup=BeautifulSoup(myUrl, 'html.parser')


# title of the page
print(soup.title)

# get attributes:
print(soup.title.name)

# get values:
print(soup.title.string)

# beginning navigation:
print(soup.title.parent.name)

# getting specific values:
print(soup.p)

# prettify() function in BeautifulSoup will enable us to view how the tags are nested in the document
print(soup.prettify())

```

#### Step 5

<ul>
    <li>Locate and scrape the services. Using <code>soup.find_all()</code> function extract the specific html element tag from the entire or specific portion of the webpage.</li>
<ul><br>
  
<example> 
<p>we should find the HTML services on this web page, extract them and store them. For each element in the web page, they would always have a unique HTML <code>"ID"</code> or <code>"class"</code>. To check their <code>ID</code> or <code>class</code>, we would need to <b>INSPECT element</b> on webpage.</p>
</example>   

```python
# soup.find_all('div') function will extract all the div tags on the given url
div_list= soup.find_all('div')

# this will be all the div tag as a single element of review[] list
print(div_list)
```

#### Step 6

<ul>
    <li>On inspecting the web page for extracting all the services name on <a href="www.eshikshak.co.in">http://www.ehsikshak.co.in</a> website, we located <code>div</code> tag with <code>class</code> value as <code>'row service-ro no-margin'</code> as parent node.
    </li>
    <li>
        To extract all the child node which is our target to extract all the services name on  <a href="www.eshikshak.co.in">http://www.ehsikshak.co.in</a> website, we locted <code>div</code> tag with <code>class</code> value as <code>'col-lg-4 col-md-6 single-servic'</code> as target node.
    </li>
</ul>

```python
# services[] a list to store all the services name on http://www.eshikshak.co.in
services=[]

# the outermost loop will help to extract all HTML element of div tag with class value as 'row service-ro no-margin'
for i in soup.find_all('div',{'class':'row service-ro no-margin'}):

    # the innnermost loop will help to extracdt all the HTML element of div tag with class value as col-lg-4 col-md-6 single-servic
    for j in i.find_all('div',{'class':'col-lg-4 col-md-6 singe-servic'}):
        # to extract h4 HTML element in each j
        per_service = j.find('h4')
        # this will print the service name 
        print(per_service.get_text())
        # to append the service name in services list
        services.append(per_service.get_text())
  
```

Above listed are the services name on <a href="http://www.eshikshak.co.in">www.eshikshak.co.in</a> website.
